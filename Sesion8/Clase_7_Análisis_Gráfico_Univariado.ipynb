```python
import seaborn as sns
import pandas as pd
```

### Gráfico de barras

- La función 'countplot' se utiliza para hacer un gráfico de barras con el conteo de las categorías, así:

```python
categorias = ["A", "A", "A", "B","B","C","C","C","C"]
sns.countplot(x = categorias)
```

Usemos la base del titanic:

```python
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
df.head(3)
```

- Gráfico de conteo de los pasajeros por clase:

```python
sns.countplot(x = df['Pclass'])
```

- La función barplot permite crear gráficos de barras en seaborn. Esta función calcula la media para cada variable categórica y la representa con barras, así como sus intervalos de confianza con barras de error.

**En este gráfico estamos calculando la proporción de personas que sobrevivieron por sexo:**

```python
sns.barplot(x = "Sex", y = "Survived", data = df)
```

**Si queremos que el anterior gráfico sea horizontal, intercambiamos las variables de eje. Note que el gráfico se adapta al tipo de dato.**

```python
sns.barplot(x = "Survived", y = "Sex", data = df)
```

**Ahora, agregemos el título:**

```python
sns.barplot(x = "Survived", y = "Sex", data = df).set(title = 'Proporción de Supervivencia por sexo')
```

Consulte la documentación de esta función en el siguiente enlace: https://seaborn.pydata.org/generated/seaborn.barplot.html

```python
# Por ejemplo, en la documentación puede leer que para que el gráfico no tenga intervalo de confianza puede
# especificar el parámetro "errorbar" así:
sns.barplot(x = "Survived", y = "Sex", data = df, errorbar=None).set(title = 'Proporción de Supervivencia por sexo')

# Note que los elementos del área del trazado se especifican dentro del paréntesis de la función, y los elementos del
# área del gráfico se especifican por fuera con la sintaxis seguida de punto.
```

**Podemos cambiar la paleta de color, con el parámetro palette**:
https://seaborn.pydata.org/tutorial/color_palettes.html

```python
sns.barplot(x = "Survived", y = "Sex", palette = "Blues_d",data = df).set(title = 'Proporción de Supervivencia por sexo')
```

### Tortas

Aquí vamos a usar **matplotlib** porque seaborn no tiene la opción para realizarlos.

- Para este gráfico con matplotlib usted, tendrá que tener los datos agrupados. Si tiene un dataframe puede utilizar la función **groupby**

```python
import matplotlib.pyplot as plt

manzanas = [20,10,25,30]
nombres = ["Ana","Juan","Diana","Catalina"]
plt.pie(manzanas, labels=nombres)
plt.axis("equal") ### ¿Que pasa si comentamos este parámetro?
plt.show()
```

- Agreguemos en la etiqueta los porcentajes

```python
plt.pie(manzanas, labels=nombres, autopct="%0.1f %%")
plt.axis("equal")
plt.show()
```

- Utilicemos la función desfase, ¿que hace?

```python
desfase = (0, 0, 0, 0.1)
plt.pie(manzanas, labels=nombres, autopct="%0.1f %%", explode=desfase)
plt.title('Participación por estudiantes')
plt.show()
```

### Boxplot - Diagrama de Caja

**Si la variable la incluimos en el parámetro x, se realiza el gráfico de manera horizontal.**

```python
sns.boxplot(x = df.Age)
```

**Si la variable la incluimos en el parámetro y, se realiza el gráfico de manera vertical.**

```python
sns.boxplot(y = df.Age)
```

**Con el parámetro color podemos cambiar de color:**

```python
sns.boxplot(y = df.Age, color = 'red')
```

**Podemos incluir más parámetros. Tamaño de los datos atípicos, grosor de las líneas, saturación del color y cambiarle el título.**

```python
sns.boxplot(y = df.Age, color = 'red',
            fliersize = 10, # Tamaño de los atípicos
            linewidth = 3,  # Grosor de las líneas
            saturation = 0.3).set(title = 'Boxpot de la edad de los pasajeros') # Saturación del color
```

**Podemos hacer un diagrama de caja incluyendo una variable categórica y mostrar el comportamiento de la edad por género:**

```python
# Incluimos las opciones de titulos de ejes en el comando set.
sns.boxplot(x = df.Sex,
            y = df.Age,
            order = ["female", "male"],
            palette = "rocket_r").set(title = 'Diagrama de caja y bigotes de la edad vs género',xlabel='Sexo', ylabel='Edad')
```

**También podemos cambiar la forma de identificar los datos atípicos (puntos por arriba o por debajo de los bigotes). Usamos el parámetro "whis", que por defecto es 1.5**

Revise la documentación de esta función para ver otros parámetros: https://seaborn.pydata.org/generated/seaborn.boxplot.html

```python
sns.boxplot(x = df.Sex,
            y = df.Age,
            order = ["female", "male"],
            palette = "rocket_r",
            whis=2).set(title = 'Diagrama de caja y bigotes de la edad vs género',xlabel='Sexo', ylabel='Edad')
```

### Violín

Muestra la distribución de datos cuantitativos a través de varios niveles de una (o más) variables categóricas de modo que esas distribuciones puedan compararse

**Si queremos cambiar la orientación de horizontal a vertical, cambiamos la variable de x a y**

```python
sns.violinplot(x=df["Age"])
```

```python
sns.violinplot(y=df["Age"])
```

**Al igual que en el Boxplot podemos agregar variables categóricas para comparar grupos de observaciones**

```python
sns.violinplot(data=df, x="Age", y="Sex")
```

**Y podemos cambiar y agregar otros parámetros:**

```python
sns.violinplot(data=df, y="Age", x="Sex", palette = "rocket_r").set(title = 'Violin de la edad vs género')
```

**Podemos agregarle, más de una variable categórica, diferenciando con color la segunda variable categórica y agregando una leyenda automáticamente:**

```python
sns.violinplot(data=df, x="Sex", y="Age", hue="Embarked", palette = "rocket_r")
```

### Histograma

Un histograma es un gráfico que se utiliza para representar la distribución de frecuencias de una variable.

```python
# Estamos incluyendo el parámetro color donde le podemos cambiar el color del gráfico.
sns.histplot(data=df, x="Age", color = 'orange')
```

**Con el parámetro binwidth podemos definir el ancho de los contenedores**

```python
sns.histplot(data=df, x="Age", binwidth=3)
```

**Con el parámetro bins podemos definir el número de contenedores**

```python
sns.histplot(data=df, x="Age", bins=17)
```

**Además, podemos agregar la densidad del kernel para suavizar el histograma con el parámetro kde**

```python
sns.histplot(data=df, x="Age", kde=True)
```

**Y podemos ver el histograma por variables categóricas**

```python
sns.histplot(data=df, x="Age", hue="Sex", palette = "viridis")
```

**Gráfico de líneas**

- Generaremos datos aleatorios para hacer una simulación

```python
import numpy as np
rng = np.random.RandomState(0)

# Simulación de datos
x = np.array(range(0, 20))
y = np.square(x) + rng.uniform(0, 100, 20)

# Conjunto de datos
df1 = {'x': x, 'y': y}
```

**lineplot, es la función para graficar lineas**

```python
sns.lineplot(x = "x", y = "y", data = df1)
```

**Utilizamos la función marker para subrayar cada punto del gráfico**

```python
sns.lineplot(x = "x", y = "y", data = df1,
             marker = "o")
```

**Con el parámetro linestyle podemos cambiar el estilo de las líneas, con color para cambiar el color y con linewidth podemos ampliar el tamaño de la línea**

```python
sns.lineplot(x = "x", y = "y", data = df1, linestyle = "dashed", color = 'red', linewidth = 3).set(title = 'Gráfico de linea')
```
