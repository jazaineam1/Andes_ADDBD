### **Algunos conceptos importantes**

**On-Premise:**
- **Almacenamiento local:** En una soluci√≥n "on-premise", los datos se almacenan en servidores y sistemas que est√°n f√≠sicamente ubicados en las instalaciones de la organizaci√≥n.
- **Control total:** La organizaci√≥n tiene un control completo sobre la infraestructura, los servidores y el entorno en el que se encuentran los datos.
- **Seguridad y privacidad:** La seguridad depende de la infraestructura y medidas de seguridad implementadas por la organizaci√≥n, lo que puede requerir inversiones significativas en t√©rminos de seguridad cibern√©tica y recursos humanos.
- **Costos iniciales:** Se requiere una inversi√≥n inicial en hardware, servidores, infraestructura y personal para gestionar y mantener los sistemas.
- **Escalabilidad limitada:** La escalabilidad puede ser m√°s complicada y costosa, ya que se deben adquirir y configurar nuevos equipos f√≠sicos a medida que crecen los requisitos de datos.
- **Acceso local:** El acceso a los datos suele limitarse a la red interna de la organizaci√≥n y puede requerir conexiones VPN para acceder de forma segura desde ubicaciones remotas.

**Data Cloud (Nube de Datos):**
- **Almacenamiento en la nube:** En una soluci√≥n de "data cloud", los datos se almacenan en servidores remotos y gestionados por proveedores de servicios en la nube.
- **Infraestructura gestionada:** La organizaci√≥n no es responsable de mantener la infraestructura f√≠sica. El proveedor de la nube se encarga de la administraci√≥n y el mantenimiento.
- **Seguridad y privacidad:** Los proveedores de la nube ofrecen medidas de seguridad robustas y certificaciones para proteger los datos. Sin embargo, la confianza en la seguridad de la nube implica ceder cierto nivel de control a terceros.
- **Costos escalonados:** Los costos suelen ser operativos en lugar de iniciales, con modelos de pago por uso. Puede ser m√°s rentable a largo plazo, especialmente para empresas que experimentan cambios en sus necesidades de almacenamiento y c√≥mputo.
- **Escalabilidad flexible:** La nube permite escalar f√°cilmente los recursos seg√∫n las necesidades cambiantes, ya que los proveedores pueden asignar recursos adicionales de manera r√°pida y eficiente.
- **Acceso remoto:** Los datos pueden accederse desde cualquier ubicaci√≥n con conexi√≥n a Internet, lo que facilita el trabajo remoto y la colaboraci√≥n.

La elecci√≥n entre "on-premise" y "data cloud" depende de los objetivos, recursos y necesidades de seguridad de una organizaci√≥n. Las soluciones "on-premise" brindan un mayor control pero requieren inversiones significativas, mientras que las soluciones en la nube ofrecen flexibilidad y escalabilidad, a expensas de ceder cierto control sobre la infraestructura a terceros.

**Miner√≠a de Datos:**

La miner√≠a de datos, tambi√©n conocida como descubrimiento de datos, es una forma especializada de an√°lisis de datos que se dirige a grandes conjuntos de datos.

En relaci√≥n con el an√°lisis de Big Data, la miner√≠a de datos generalmente se refiere a t√©cnicas automatizadas basadas en software que tamizan conjuntos de datos masivos para identificar patrones y tendencias.

La miner√≠a de datos constituye la base para anal√≠tica predictiva e inteligencia empresarial (BI).

**An√°lisis Estad√≠stico:**

<img src="https://cdn-icons-png.flaticon.com/512/786/786695.png" alt="Mi imagen" width="230">

 El an√°lisis estad√≠stico utiliza m√©todos estad√≠sticos basados ‚Äã‚Äãen f√≥rmulas matem√°ticas como medio para analizar datos.Este tipo de an√°lisis se usa com√∫nmente para describir conjuntos de datos a trav√©s de res√∫menes, como proporcionando la media, la mediana o la moda de las estad√≠sticas asociadas con el conjunto de datos.

Tambi√©n puede utilizarse para inferir patrones y relaciones dentro del conjunto de datos, como regresi√≥n y correlaci√≥n.

* Pruebas A/B: Las pruebas A/B, tambi√©n conocidas como pruebas divididas o de cubo, comparan dos versiones de un elemento para determinar qu√© versi√≥n es superior en funci√≥n de una m√©trica predefinida.

>Por ejemplo, para determinar el mejor dise√±o posible para un anuncio de helado en el sitio web de la empresa A, se utilizan dos versiones diferentes del anuncio. La versi√≥n A es una existente ad (el control) mientras que la Versi√≥n B ha tenido su dise√±o ligeramente alterado (el tratamiento). Luego, las versiones se muestran simult√°neamente a diferentes usuarios:

 >Versi√≥n A al Grupo A

 >Versi√≥n B al Grupo B

 * Correlaci√≥n: La correlaci√≥n es una t√©cnica de an√°lisis utilizada para determinar si dos variables est√°n relacionadas entre s√≠.

>La correlaci√≥n es por lo tanto com√∫nmente utilizado para la miner√≠a de datos donde la identificaci√≥n de las relaciones entre variables en un conjunto de datos conduce al descubrimiento de patrones y anomal√≠as.

>Por ejemplo, los gerentes creen que las helader√≠as necesitan almacenar m√°s helados para los clientes en d√≠as calientes, pero no s√© cu√°nto m√°s almacenar.

* Regresi√≥n: La t√©cnica de an√°lisis de regresi√≥n explora c√≥mo una variable dependiente se relaciona con una variable independiente dentro de un conjunto de datos.

>La aplicaci√≥n de esta t√©cnica ayuda a determinar c√≥mo cambia el valor de la variable dependiente en relaci√≥n con los cambios en el valor de la variable independiente.

>Por ejemplo, para determinar cu√°nto stock adicional necesita cada helader√≠a, los analistas aplican la regresi√≥n ingresando los valores de las lecturas de temperatura. Estos se basan en el pron√≥stico del tiempo como variable independiente y la cantidad de cremas vendidas como variable dependiente. Lo que descubren los analistas es que el 15% de se requiere stock adicional por cada aumento de temperatura de 5 grados.

* Calsificaci√≥n Supervisado:  La clasificaci√≥n es una t√©cnica de aprendizaje supervisado mediante la cual los datos se clasifican en relevantes, categor√≠as previamente aprendidas. Consiste en dos pasos:

>El sistema recibe datos de entrenamiento que ya est√°n categorizados o etiquetados, para que pueda desarrollar una comprensi√≥n de las diferentes categor√≠as.

>El sistema recibe datos desconocidos pero similares para la clasificaci√≥n y se basa en la entendi√©ndolo desarrollado a partir de los datos de entrenamiento, el algoritmo clasificar√° los datos sin etiquetar.

* Calsificaci√≥n No  Supervisado: El agrupamiento es una t√©cnica de aprendizaje no supervisado mediante la cual los datos se dividen en diferentes grupos para que los datos de cada grupo tengan propiedades similares.

>No hay un aprendizaje previo de categor√≠as requeridas.

>La agrupaci√≥n en cl√∫steres se usa generalmente en la miner√≠a de datos para obtener una comprensi√≥n de las propiedades de un conjunto de datos dado.

>Despu√©s de desarrollar esta comprensi√≥n, la clasificaci√≥n se puede utilizar para hacermejores predicciones sobre datos similares pero nuevos o no vistos.
