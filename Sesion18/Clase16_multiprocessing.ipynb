Un sitemap de una página web es básicamente un mapa o una guía que enumera todas las páginas importantes de un sitio web. Se puede ver  como un índice o tabla de contenidos que le dice a los motores de búsqueda y a los visitantes dónde está cada pieza de contenido en tu sitio. Los sitemaps son especialmente útiles para sitios web grandes, con muchas páginas, o para aquellos con contenido que no está fácilmente accesible a través de enlaces.

### Propósitos y Beneficios de un Sitemap

1. **Mejora la Indexación por Motores de Búsqueda**:
   - Los sitemaps son una herramienta crucial para los motores de búsqueda como Google. Ayudan a los rastreadores de estos motores a encontrar y entender el contenido de tu sitio web. Cuando envías un sitemap a un motor de búsqueda, básicamente le estás diciendo: "Aquí están las páginas de mi sitio web, por favor inclúyelas en tu índice".

2. **Facilita la Navegación**:
   - Para los visitantes del sitio, un sitemap puede actuar como una herramienta de navegación rápida, permitiéndoles encontrar la información que buscan sin tener que navegar por todo el sitio.

3. **Organización y Estructura del Sitio Web**:
   - Crear un sitemap te obliga a pensar en la estructura y organización de tu sitio, lo cual es crucial para una buena experiencia de usuario y SEO (optimización para motores de búsqueda).

4. **Identificación de URLs y Contenido**:
   - Los sitemaps permiten a los motores de búsqueda identificar rápidamente las URLs de tu sitio y cualquier contenido actualizado, lo cual es esencial para sitios con contenido que cambia frecuentemente.

### Tipos de Sitemaps

1. **Sitemap XML**:
   - Es el tipo más común de sitemap. Está formateado de manera que los motores de búsqueda pueden leer e interpretar fácilmente el archivo. Los sitemaps XML listan las URLs de un sitio y pueden incluir información adicional como la última fecha de actualización, la frecuencia de cambios y la importancia de las URLs.

```python
import concurrent.futures
import time
import random

# Función que simula la descarga de contenido de una URL
def download_content(url):
    # Simula la duración de la descarga
    download_duration = 10
    print(f"Descargando desde {url} (duración: {download_duration} segundos)")
    time.sleep(download_duration)
    return f"Contenido de {url} descargado."

# Lista de URLs ficticias
urls = [
    "http://sitio1.com/data",
    "http://sitio2.com/data",
    "http://sitio3.com/data",
    "http://sitio4.com/data",
    "http://sitio5.com/data"
]

# Usamos ThreadPoolExecutor para descargar el contenido de las URLs en paralelo
with concurrent.futures.ThreadPoolExecutor() as executor:
    # Mapeamos la función 'download_content' a cada URL
    results = list(executor.map(download_content, urls))

# Imprimimos los resultados
for result in results:
    print(result)
```

El método `ThreadPoolExecutor.map` del módulo `concurrent.futures` en Python es una herramienta poderosa para ejecutar operaciones en paralelo. Es especialmente útil cuando tienes una función que quieres ejecutar con diferentes argumentos de manera concurrente. Aquí te doy una clase sobre cómo funciona `ThreadPoolExecutor.map` y cómo se usa en el contexto de tu script:

### 1. ¿Qué es `ThreadPoolExecutor`?

`ThreadPoolExecutor` es una clase en el módulo `concurrent.futures` que proporciona una interfaz de alto nivel para ejecutar llamadas a funciones de manera asincrónica (es decir, de forma paralela o concurrente). Utiliza un pool de hilos (threads) para ejecutar las llamadas a funciones, lo que permite realizar múltiples operaciones al mismo tiempo.

### 2. ¿Qué hace el método `.map`?

El método `.map` es similar al `map` integrado de Python, pero está diseñado para ejecutar las funciones en paralelo utilizando el pool de hilos de `ThreadPoolExecutor`. Lo que hace es tomar una función y un iterable (como una lista), y luego aplicar esa función a cada elemento del iterable. La gran ventaja aquí es que cada llamada a la función se ejecuta en un hilo separado, lo que permite que las operaciones se realicen de manera concurrente.

### 3. ¿Cómo funciona en tu script?

En tu script, `ThreadPoolExecutor.map` se utiliza para aplicar la función `download_content` a cada URL en la lista `urls`. Esto es lo que sucede:

- **Inicialización**: `ThreadPoolExecutor` se inicializa utilizando el bloque `with`, lo que garantiza que todos los hilos se cierren correctamente al finalizar las operaciones.

- **Aplicación de la Función**: `executor.map(download_content, urls)` aplica `download_content` a cada elemento en `urls`. Bajo el capó, `ThreadPoolExecutor` asigna cada llamada a la función a un hilo diferente en su pool de hilos.

- **Ejecución Concurrente**: Las funciones se ejecutan en paralelo. Si tienes, por ejemplo, 5 URLs y un pool de 5 hilos, las 5 descargas podrían, en teoría, suceder al mismo tiempo.

- **Recopilación de Resultados**: `executor.map` devuelve un iterador que se convierte en una lista. Los resultados se devuelven en el orden en que se pasaron las tareas, es decir, el primer resultado corresponde al primer elemento en `urls`, y así sucesivamente.

- **Impresión de Resultados**: Al final, imprimes los resultados, que son los valores devueltos por `download_content` para cada URL.

### 4. Ventajas de usar `ThreadPoolExecutor.map`

- **Concurrencia**: Permite realizar múltiples operaciones de I/O (como solicitudes de red) de manera más eficiente, ya que no estás esperando a que cada operación se complete antes de iniciar la siguiente.
  
- **Facilidad de Uso**: Proporciona una forma sencilla de convertir un bucle que procesa elementos de manera secuencial en un conjunto de operaciones concurrentes, con mínimos cambios en el código.

- **Gestión de Recursos**: El manejo de los hilos y su ciclo de vida se gestionan automáticamente, lo que reduce el riesgo de errores relacionados con la concurrencia.

### Tokenización
La tokenización es un paso fundamental en el procesamiento del lenguaje natural (NLP) y el aprendizaje automático. Consiste en dividir un texto en unidades más pequeñas, llamadas "tokens". Estos tokens pueden ser palabras, subpalabras, caracteres o incluso grupos de palabras, dependiendo del tipo de tokenización que se aplique. El objetivo principal de la tokenización es transformar el texto en una forma que pueda ser entendida y procesada por algoritmos y modelos de aprendizaje automático.

```python
def concatenate_text(examples):
    return {
        "text": examples["Titulo"]
        + " \n "
        + examples["Contenido"]

    }


comments_dataset = comments_dataset.map(concatenate_text)
```

```python
from transformers import AutoTokenizer, TFAutoModel

model_ckpt = "sentence-transformers/multi-qa-mpnet-base-dot-v1"
tokenizer = AutoTokenizer.from_pretrained(model_ckpt)
model = TFAutoModel.from_pretrained(model_ckpt, from_pt=True)
```

```python
def cls_pooling(model_output):
    return model_output.last_hidden_state[:, 0]
```

```python
max_length = 512  # Límite máximo del modelo


# Función de tokenización y verificación de truncamiento
def get_embeddings(text_list):
    encoded_input = tokenizer(
        text_list, padding=True, truncation=True, max_length=max_length, return_tensors="tf"
    )
    truncated = [len(tokens) == max_length for tokens in encoded_input['input_ids']]
    encoded_input = {k: v for k, v in encoded_input.items()}
    model_output = model(**encoded_input)
    return cls_pooling(model_output), truncated
```

```python
def map_embeddings(example):
    embeddings, truncated = get_embeddings([example["text"]])
    return {"embeddings": embeddings.numpy()[0], "truncated": truncated[0]}
```

```python
embeddings_dataset = comments_dataset.map(map_embeddings)
embeddings_dataset
```

```python
filtered_dataset = embeddings_dataset.filter(lambda x: not x["truncated"])
filtered_dataset
```

## FAISS: Facebook AI Similarity Search
FAISS (Facebook AI Similarity Search) es una biblioteca de código abierto desarrollada por Facebook AI Research (FAIR) que permite la búsqueda rápida y eficiente de similitudes en grandes conjuntos de datos vectoriales. Es particularmente útil en aplicaciones de búsqueda y recuperación de información, como motores de búsqueda, sistemas de recomendación, y recuperación de imágenes.

```python
embeddings_dataset.add_faiss_index(column="embeddings")
```

```python
# Función para obtener el embedding de una pregunta
def get_question_embedding(question):
    embeddings, _ = get_embeddings([question])  # Ignora la información de truncamiento
    return embeddings.numpy()

```

```python
# Obtener el embedding de la pregunta
question = "futbolistas importantes"
question_embedding = get_question_embedding(question)
print(question_embedding.shape)
```

```python

# Obtener los ejemplos más cercanos
scores, samples = embeddings_dataset.get_nearest_examples(
    "embeddings", question_embedding, k=10
)
```

```python
samples_df = pd.DataFrame.from_dict(samples)
samples_df["scores"] = scores
samples_df.sort_values("scores", ascending=True, inplace=True)
```

```python
for _, row in samples_df.iterrows():
    print(f"SCORE: {row.scores}")
    print(f"TITLE: {row.Titulo}")
    print(f"URL: {row.Url}")
    print("=" * 50)
    print()
```
